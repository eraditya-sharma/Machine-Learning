```python
#Encoding the gender with Binary Encoding (Explicit Type Casting)
x_ghw.Gender[x_ghw.Gender=='Female']=0
x_ghw.Gender[x_ghw.Gender=='Male']=1
```

```python
#Reshaping the input and output to array
xx_ghw = x_ghw.values.reshape(500,3)
yy_index = y_index.values.reshape(500,1)
#Note: Change the variable as if the reshaped x_ghw will be stored in x_ghw then on first entry it will convert to 
```

```python
#Segregating the data into 70:30 ratio of train:test datasets
x_train = xx_ghw[:350]
x_test = xx_ghw[350:]

y_train = yy_index[:350]
y_test = yy_index[350:]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
trainer = KNeighborsClassifier()    #by default value of K is 5
```

```python
learner = trainer.fit(x_train, y_train)
```
